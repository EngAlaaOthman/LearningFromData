**Goal:** Demonstrate that with more training data, the out-of-sample error ($E_{out}$) converges to the in-sample error ($E_{in}$), as predicted by Hoeffding's Inequality.

## What You Will Do

1. For different training set sizes $N = [10, 20, 50, 100, 200, 500]$:
   - Train PLA on $N$ random points.
   - Compute $E_{in}$ (training error).
   - Compute $E_{out}$ (test error on 1000 fresh points).
   - Repeat **10 times** for each $N$ (to get reliable averages).
3. Plot the results:
   - Average $E_{in}$ and $E_{out}$ vs. $N$.
   - Generalization gap $|E_{out} - E_{in}|$ vs. $N$.
   - Overlay the Hoeffding bound $2e^{-2\epsilon^2 N}$ for $\epsilon = 0.1$.

## Questions to Answer

1. Does the gap $|E_{out} - E_{in}|$ shrink as $N$ increases?
2. At what $N$ does the Hoeffding bound become $\leq 0.05$?
3. Does the actual gap stay below the bound? Why or why not?
